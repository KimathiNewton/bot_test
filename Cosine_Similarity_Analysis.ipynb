## Analysing Olivia Responses


```python
olivia_df=pd.read_csv("data/olivia_conversation_with_cosine.csv")
olivia_df.head()
```

```python
olivia_df.describe()
```

```python
# Filter Bella's low-scoring responses
low_scoring_olivia = olivia_df[olivia_df['cosine_olivia'] < 0.8]

# Display the low-scoring responses for analysis
low_scoring_olivia[['Question', 'Olivia_Target_Response', 'Olivia_Response', 'cosine_olivia']]
```

For questions about age and work experience, the responses are factually correct and relevant, with slight differences in detail. However, the cosine similarity score for the age question is unjustifiably low (0.369)

In questions about marital status and profession, Olivia's responses provide either uncertainty (marital status) or more specific details (profession) compared to the general target responses. While the core meaning is somewhat aligned, these discrepancies impact the cosine similarity scores. Scores in the 0.7-0.8 range seem slightly inflated for these cases, as the responses don’t match perfectly.

In the parental status question, there’s a direct factual mismatch (no information vs. yes). This justifies the lower similarity score (0.578). This response is the least aligned with the target.

Even when there are discrepancies, many of Olivia’s responses remain relevant, though they sometimes offer additional details or avoid giving a direct answer (e.g., uncertainty in marital status). This indicates that the bot is capable of forming related responses, but not always exact ones.


```python
# Filter Bella's low-scoring responses
high_scoring_olivia = olivia_df[olivia_df['cosine_olivia'] ==1]

# Display the high-scoring responses for analysis
high_scoring_olivia[['Question', 'Olivia_Target_Response', 'Olivia_Response', 'cosine_olivia']]
```

All the responses that received a perfect score are either identical or exhibit no differences in meaning, making the perfect cosine similarity score completely appropriate.
Olivia's responses in these cases are completely aligned with the target responses.
